```python
import pandas as pd
import numpy as np
```

```python
mm1 = pd.read_csv('playing_memory_ep30.csv')
mm2 = pd.read_csv('playing_memory_ep60.csv')
mm3 = pd.read_csv('playing_memory_ep100.csv')
mm4 = pd.read_csv('playing_memory_ep140.csv')
mm5 = pd.read_csv('playing_memory_ep180.csv')
mm6 = pd.read_csv('playing_memory_ep200.csv')
```

```python
mm.to_csv('playing_memory_200.csv', index=False)
```

```python
mm240 = pd.read_csv('playing_memory_ep240.csv')
mm280 = pd.read_csv('playing_memory_ep280.csv')
mm320 = pd.read_csv('playing_memory_ep320.csv')
mm360 = pd.read_csv('playing_memory_ep360.csv')
mm400 = pd.read_csv('playing_memory_ep400.csv')
```
